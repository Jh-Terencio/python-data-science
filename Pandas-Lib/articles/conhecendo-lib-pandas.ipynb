## Introdução
A Ciência de Dados é um ramo que vem ganhando cada vez mais notoriedade ao redor do mundo. Nesse cenário, a linguagem Python é bastante utilizada devido à sua versatilidade e simplicidade, contando com uma vasta quantidade de bibliotecas, entre elas, o Pandas, uma das ferramentas essenciais quando se fala em Ciência de Dados.

## O que é Pandas
Pandas é uma biblioteca para Ciência de Dados que providencia uma abordagem rápida e flexível, com estruturas robustas para se trabalhar com dados relacionais (ou rotulados).

### Utilizações
- Limpeza e tratamento de dados
- Análise exploratória de dados (EDA)
- Webscraping

## Como funciona o Pandas
Dentro do pacote Pandas, temos dois objetos primários importantes: as Series e os DataFrames.

### Series `pandas.Series`
As Series são objetos de tipo array unidimensional, com um eixo de rótulos, também chamado de index, que é responsável por identificar cada registro.

### DataFrame `pandas.DataFrame`
Os DataFrames são objetos bidimensionais, de tamanho variável. O seu formato é de uma tabela, onde os dados são organizados em linhas e colunas.
